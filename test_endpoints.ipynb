You can test your blockchain user content API endpoints using various methods. Here are the most common approaches:

## 1. Using Postman

Postman is a popular tool for API testing with a user-friendly interface:

1. **Download and install** [Postman](https://www.postman.com/downloads/)
2. **Create a new collection** for your blockchain API tests
3. **Set up requests** for each endpoint:

### GET User Example
- Method: GET
- URL: `http://localhost:3000/api/user/0x1234567890abcdef`

### GET Content Example
- Method: GET
- URL: `http://localhost:3000/api/content/content123`

### POST Update User Example
- Method: POST
- URL: `http://localhost:3000/api/user/update`
- Body (raw JSON):
  ```json
  {
    "user_address": "0x1234567890abcdef",
    "field": "total_followers",
    "value": 5
  }
  ```

### POST Update Content Example
- Method: POST
- URL: `http://localhost:3000/api/content/update`
- Body (raw JSON):
  ```json
  {
    "content_id": "content123",
    "action": "add_comment",
    "data": {
      "user_address": "0x1234567890abcdef",
      "text": "This is a test comment"
    }
  }
  ```

## 2. Using cURL

If you prefer command line testing:

### GET User
```bash
curl -X GET http://localhost:3000/api/user/0x1234567890abcdef
```

### GET Content
```bash
curl -X GET http://localhost:3000/api/content/content123
```

### POST Update User
```bash
curl -X POST http://localhost:3000/api/user/update \
  -H "Content-Type: application/json" \
  -d '{"user_address":"0x1234567890abcdef","field":"total_followers","value":5}'
```

### POST Update Content
```bash
curl -X POST http://localhost:3000/api/content/update \
  -H "Content-Type: application/json" \
  -d '{"content_id":"content123","action":"add_like","data":{"user_address":"0x1234567890abcdef"}}'
```

## 3. Writing Automated Tests

For more robust testing, you can create automated tests using Jest or Mocha:

1. Install testing packages:
   ```bash
   npm install --save-dev jest supertest
   ```

2. Create a test file (e.g., `tests/api.test.js`)

3. Add the following to your `package.json`:
   ```json
   "scripts": {
     "test": "jest"
   }
   ```

4. Run tests with `npm test`

## Testing Workflow

1. Start your API server: `npm run dev`
2. Test creating new users and content
3. Test updating users and content
4. Verify data is correctly stored in MongoDB (use MongoDB Compass to inspect)
5. Test error cases (invalid inputs, non-existent resources)
